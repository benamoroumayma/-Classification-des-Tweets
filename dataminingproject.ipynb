## <font color=#ff751a style="text-decoration: underline;">1 √©re Technique:</font>
* Etant donn√© un ensemble de tweets, l‚Äôobjectif est de les r√©sumer sous formes de groupes de sorte <br>√† ce que les Tweets qui sont dans le m√™me groupe soient similaires. Ainsi, l‚Äôutilisateur pourra par lasuite<br> lire juste un Tweet de chaque groupe (le Tweet qui est le centro√Øde de groupes).

### <font color=#ff471a> K-means clustering: </font>

* M√©thode du code pour s√©lectionner le nombre de clusters
* Cette m√©thode examine le pourcentage de variance expliqu√© en fonction du nombre de clusters: il faut choisir un certain nombre de clusters afin que l'ajout d'un autre cluster ne donne pas une meilleure mod√©lisation des donn√©es. Plus pr√©cis√©ment, si l'on trace le pourcentage de variance expliqu√© par les clusters par rapport au nombre de clusters, les premiers clusters ajouteront beaucoup d'informations (expliquent beaucoup de variance), mais √† un moment donn√© le gain marginal diminuera, donnant un angle dans le graphique. Le nombre de clusters est choisi √† ce stade, d'o√π le "crit√®re du coude". 
* Ce ¬´coude¬ª ne peut pas toujours √™tre identifi√© sans ambigu√Øt√©. Le pourcentage de variance expliqu√© est le rapport de la variance entre les groupes √† la variance totale, √©galement appel√© test F. Une l√©g√®re variation de cette m√©thode trace la courbure de la variance intra-groupe.

```python
tfidf_vectorizer = TfidfVectorizer(max_df=0.90, min_df=2, max_features=1000, stop_words='english')
# TF-IDF feature matrix
v = tfidf_vectorizer.fit_transform(fichierTweet['Tweet'].values.astype('U'))
```

```python
print(v)
```

```python
from sklearn.cluster import KMeans
w = []
for i in range(1,22):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(v)
    w.append(kmeans.inertia_)
```

```python
plt.plot(range(1,22),w)
plt.title('cluster')
plt.xlabel('Nombre de clusters')
plt.ylabel('W')
plt.show()
```

* Comme plus d'un coudes ont √©t√© g√©n√©r√©s, je devrai s√©lectionner la bonne quantit√© de grappes par essais et erreurs. Donc, je vais pr√©senter les r√©sultats de diff√©rentes quantit√©s de grappes pour trouver la bonne quantit√© de grappes.

```python
kmeans = KMeans(n_clusters = 22, n_init = 20, n_jobs = 1) # n_init(number of iterations for clsutering) n_jobs(number of cpu cores to use)
kmeans.fit(v)
```

```python
# We look at 22 the clusters generated by k-means.
common_words = kmeans.cluster_centers_.argsort()[:,-1:-26:-1]
for num, centroid in enumerate(common_words):
    print(str(num) + ' : ' + ', '.join(str(fichierTweet['Tweet'] )for Tweet in centroid))
   
```

### <font color=#ff471a>1. Similitude Jaccard : </font>
* La similarit√© Jaccard ou l'intersection sur l'union est d√©finie comme la taille de l'intersection divis√©e par la taille de l'union de deux ensembles.
* La similarit√© Jaccard prend uniquement un ensemble unique de mots pour chaque phrase ou document, tandis que la similarit√© cosinus prend la longueur totale des vecteurs. La similitude Jaccard est bonne pour les cas o√π la duplication n'a pas d'importance, la similitude cosinus est bonne pour les cas o√π la duplication est importante. Dans notre cas, le contexte compte plus que la duplication, donc la similarit√© Jaccard est la technique id√©ale √† utiliser.

### <font color=#6666ff style="text-decoration: underline;">Scores de similarit√© Jaccard

```python
'''Jaccard similarity is good for cases where duplication does not matter, 
cosine similarity is good for cases where duplication matters while analyzing text similarity. For two product descriptions, 
it will be better to use Jaccard similarity as repetition of a word does not reduce their similarity.'''

def jaccard_similarity(query, document):
    intersection = set(query).intersection(set(document))
    union = set(query).union(set(document))
    return len(intersection)/len(union)
# jaccard_score(socialvector, economic_vector)

#for similarity of 1 and 2 of column1
# jaccard_similarity('dog lion a dog','dog is cat')


def get_scores(group,tweets):
    scores = []
    for tweet in tweets:
        s = jaccard_similarity(group, tweet)
        scores.append(s)
    return scores
```

```python
'''Actual assigning of classes to the tweets'''

def get_clusters(l1, l2, l3, l4):
    econ = []
    socio = []
    cul = []
    heal = []
    for i, j, k, l in zip(l1, l2, l3, l4):
        m = max(i, j, k, l)
        if m == i:
            econ.append(1)
        else:
            econ.append(0)
        if m == j:
            socio.append(1)
        else:
            socio.append(0)        
        if m == k:
            cul.append(1)
        else:
            cul.append(0)  
        if m == l:
            heal.append(1)
        else:
            heal.append(0)   
            
    return econ, socio, cul, heal
```

```python
l1 = scores_df.economic_score.to_list()
l2 = scores_df.social_score.to_list()
l3 = scores_df.culture_score.to_list()
l4 = scores_df.health_scores.to_list()

econ, socio, cul, heal = get_clusters(l1, l2, l3, l4)
```

```python

data = {'name': scores_df.names.to_list(), 'economic':econ, 'social':socio, 'culture':cul, 'health': heal}
cluster_df = pd.DataFrame(data)
cluster_df.head(5)
```
